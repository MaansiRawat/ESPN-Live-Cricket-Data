# ESPNCRCIFO LIVE DATA ANALYSIS:

```python
import pandas as pd
```

```python
df=pd.read_html('https://www.espncricinfo.com/rankings/content/page/211271.html')
```

```python
df
```

# CREATE DATAFRAME:

```python
data1=df[0]
```

```python
data1
```

```python
data2=df[1]
```

```python
data2
```

```python
data3=df[2]
```

```python
data3
```

```python
data4=df[3]
```

```python
data4
```

```python
data5=df[4]
```

```python
data5
```

# DATAFRAME MERGE:

```python
new_data=pd.concat([data1,data2,data3,data4,data5],ignore_index=True)
```

# GIVE RANGE TO POS:

```python
new_data.Pos=range(1,187)
```

# CREATE NEW CSV FILE:

```python
new_data.to_csv("Cricket.csv",index=False)
```
